#### Lambda

A função lambida é chamada de função anonima, ela se escreva (lambda x : f(x)) onde x é uma variavel anomiva e f(x) é uma função que utiliza essa variável, segue os exemplos

````python
a = (lambda x : x**2)(2)
print(a)
a = (lambda x,y : x * y)(2,3)
print(a)
potencia = lambda x,y : x**y
print(potencia(5,2))
````
